```python
new_df['genre'] = new_df['genre'].apply(lambda x:[i.replace(" ", "")for i in x])
```

```python
o = new_df[['name', 'genre','rating', 'members']]
o.info()
```

```python
o['genre'] = o['genre'].apply(lambda x:" ".join(x))
```

```python
o['genre'] = o['genre'].apply(lambda x:x.lower())
```

```python
o
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def helper(s):
    l = []
    for i in s.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
o["genre"] = o["genre"].apply(helper)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = cv.fit_transform(o["genre"]).toarray()
```

```python
k = cv.get_feature_names_out()
for i in k:
    print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend5animes(s):
    idx = o[o["name"]==s].index[0]
    l = []

    for i in range(len(similarity)):
        # Calculate weighted similarity based on members
        weighted_similarity = similarity[idx][i] * (o["members"].iloc[idx] + o["members"].iloc[i]) / 2

        l.append((weighted_similarity, o["rating"].iloc[i], i))

    l.sort(reverse=True)
    new = []

    for i,j,k in l[1:6]:
        new.append(o["name"].iloc[k])

    return new
```

# Interview Question :

## 1. Can you explain the difference between user-based and item-based collaborative filtering?


#### User-Based Collaborative Filtering (UBCF)

- Focus: Finds similar users based on their historical ratings or preferences.   

- Recommendation: Recommends items that similar users have liked.   

- Challenge: Requires a large number of users with similar tastes to make accurate recommendations.
   
#### Item-Based Collaborative Filtering (IBCF)

- Focus: Finds similar items based on how users have rated them.   

- Recommendation: Recommends items that are similar to items a user has liked.   

- Challenge: Requires a large number of ratings for each item to calculate similarity.

# 2. What is collaborative filtering, and how does it work?

Collaborative Filtering

- Personalized Recommendations: Suggests items based on user preferences and behavior.

- User-Based: Finds similar users and recommends items they liked.

- Item-Based: Finds similar items and recommends items similar to the user's past preferences.

- Widely Used: In various applications like e-commerce, streaming services, and social media.

How Collaborative Filtering Works:

1. Data Collection: Collects user ratings or interactions with items (e.g., likes, purchases, clicks). 

  
2. Similarity Calculation:

- User-Based: Finds users with similar preferences by comparing their ratings or interactions. 
  
- Item-Based: Finds similar items based on how users have rated them.   

3. Prediction: Predicts a user's rating for an item by considering the ratings of similar users or items.   

4. Recommendation: Recommends items with the highest predicted ratings to the user.

Example: If a user frequently watches sci-fi movies and has rated them highly, a collaborative filtering system might recommend other sci-fi movies that similar users have enjoyed.
