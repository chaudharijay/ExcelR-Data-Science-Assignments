# Scaling Of Data

### 1.Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Select only numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns
df_numerical = df[numerical_columns]

# Apply standardization to the numerical data
std_scaler = StandardScaler()
std_data = std_scaler.fit_transform(df_numerical)
```

```python
std_data
```

### when we need features values with mean zero and one standard deviation.In simple words when the range of values should be in between -3 to +3 we use standardization.

### 2. Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = scaler.fit_transform(df_numerical)
```

```python
data
```

### when we need feature scaling values in between 0 to 1 we use normalization.

# 2.Encoding Techniques

### one hot encoding

```python
df.columns
```

```python
df.info()
```

```python
df['workclass'] = df['workclass'].astype('category')
df['education'] = df['education'].astype('category')
df['marital_status'] = df['marital_status'].astype('category')
df['occupation'] = df['occupation'].astype('category')
```

```python
df['workclass'] = df['workclass'].cat.codes
df['education'] = df['education'].cat.codes
df['marital_status'] = df['marital_status'].cat.codes
df['occupation'] = df['occupation'].cat.codes
```

One-Hot Encoding generates several binary columns according to the number of distinct categories.

```python
df.info()
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = df.copy()
labelencoder = LabelEncoder()
df1.iloc[:,-1] = labelencoder.fit_transform(df1.iloc[:,-1])
df1
```

Label Encoding only generates a single column for the categorical variable.

Feature Engineering

```python
df['age1'] = pd.cut(df['age'],bins = [24,40,65,93],labels=['youngster','adult','senior'])

df['hours_per_week1'] = pd.cut(df['hours_per_week'],bins = [20,40,60,80],labels = ['fulltime','parttime','overtime'])

print(df[['age1','hours_per_week1']])
```

Feature Selection:

Isolation forest

```python
from sklearn.ensemble import IsolationForest
clf = IsolationForest(random_state = 34,contamination=0.01)
df_encoded = pd.get_dummies(df)
clf.fit(df_encoded)
```

```python
y_pred_outliers = clf.predict(df_encoded)

y_pred_outliers
```

```python
df_encoded
```

```python

outliers = clf.fit_predict(df.select_dtypes(include=['int64', 'float64']))
df_cleaned = df[outliers == 1]


print(f"Original shape: {df.shape}, Cleaned shape: {df_cleaned.shape}")
```
