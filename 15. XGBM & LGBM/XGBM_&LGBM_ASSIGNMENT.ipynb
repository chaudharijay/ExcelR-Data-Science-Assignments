## Data Preprocessing

```python
df.isnull().sum()
```

```python
df['Age'].fillna(0,inplace=True)
df['Cabin'].fillna(0,inplace=True)
df['Embarked'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

### Encoding

```python
df=df.drop(columns = ["Name","Cabin"])
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Ticket']=le.fit_transform(df['Ticket'])
df['Sex']=le.fit_transform(df['Sex'])

```

```python
df['Embarked'] = df['Embarked'].astype(str)
df['Embarked']=le.fit_transform(df['Embarked'])
```

### Scaling

```python
#standardization
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
std_data = std_scaler.fit_transform(df)
std_data
```

```python
#normalization
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
norm_data = minmax.fit_transform(df)
norm_data
```

## Building Predictive Models

```python
df.columns
```

```python
#train test split
from sklearn.model_selection import train_test_split
y=df['Survived']
x=df.drop(columns=['Survived'])
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=36)
```

### Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
acc = lr.score(x_test,y_test)
print("logisticregression accuracy is :",acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix
clf = RandomForestClassifier(n_estimators=100)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

### XGBOOST Model

```python
from numpy import loadtxt
from xgboost import XGBClassifier
#fitting model
model = XGBClassifier()
model.fit(x_train,y_train)
#finding predictions
y_pred = model.predict(x_test)
predictions=[round(value) for value in y_pred]

```

```python
#accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,predictions)
print("Accuracy: %.2f%%"%(accuracy*100.0))
```

```python
from sklearn.model_selection import cross_val_score
model = XGBClassifier(n_estimators=100,max_depth=5,max_leaves=10)
results = cross_val_score(model,x,y)
results
```

### LightGBM

```python
import lightgbm as lgb
l_train = lgb.Dataset(x_train,label=y_train)
```

```python
params = {}
params['learning_rate'] = 0.003
params['boosting_type'] = 'gbdt'
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5
params['num_leaves'] = 10
params['min_data'] = 50
params['max_depth'] = 10
```

```python
clf = lgb.train(params,l_train,100)
y_pred = clf.predict(x_test)
```

```python
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test,predictions)
accuracy
```
