```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
data = pd.read_csv('wine.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.hist(bins=15, figsize=(15, 10))
plt.suptitle("Histograms of Features")
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=data)
plt.title("Boxplot of Features")
plt.xticks(rotation=90)
plt.show()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
pca = PCA()
pca_data = pca.fit_transform(data_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```

```python
n_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Optimal number of components to retain 95% variance: {n_components}")
```

```python
pca = PCA(n_components=n_components)
pca_data_reduced = pca.fit_transform(data_scaled)
```

```python
kmeans_original = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans_original.fit_predict(data_scaled)
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data.iloc[:, 0], y=data.iloc[:, 1], hue=clusters_original, palette='viridis')
plt.title("Clustering on Original Data")
plt.show()
```

```python
silhouette_original = silhouette_score(data_scaled, clusters_original)
davies_bouldin_original = davies_bouldin_score(data_scaled, clusters_original)
print(f"Silhouette Score (Original Data): {silhouette_original}")
print(f"Davies-Bouldin Index (Original Data): {davies_bouldin_original}")
```

```python
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(pca_data_reduced)
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=pca_data_reduced[:, 0], y=pca_data_reduced[:, 1], hue=clusters_pca, palette='viridis')
plt.title("Clustering on PCA-Reduced Data")
plt.show()
```

```python
silhouette_pca = silhouette_score(pca_data_reduced, clusters_pca)
davies_bouldin_pca = davies_bouldin_score(pca_data_reduced, clusters_pca)
print(f"Silhouette Score (PCA Data): {silhouette_pca}")
print(f"Davies-Bouldin Index (PCA Data): {davies_bouldin_pca}")
```

```python
print("\nComparison of Clustering Performance:")
print(f"Silhouette Score - Original: {silhouette_original}, PCA: {silhouette_pca}")
print(f"Davies-Bouldin Index - Original: {davies_bouldin_original}, PCA: {davies_bouldin_pca}")
```

```python
print("\nKey Insights:")
print(f"PCA retained {n_components} components while preserving 95% variance.")
print("Clustering performance comparison shows that PCA can simplify datasets without significantly reducing clustering quality.")
print("PCA is recommended for high-dimensional data to reduce computational complexity and improve interpretability.")
```
